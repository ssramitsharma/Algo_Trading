```python
from __future__ import annotations

import os
from pathlib import Path

# Make matplotlib + font cache writable (avoids slow temp-cache rebuilds)
cache_dir = Path.cwd() / ".cache" / "matplotlib"
cache_dir.mkdir(parents=True, exist_ok=True)
os.environ.setdefault("MPLCONFIGDIR", str(cache_dir))

import gymnasium as gym
import gym_anytrading
import numpy as np
import pandas as pd
import yfinance as yf
from stable_baselines3 import DQN


def get_price_data(
    ticker: str = "NVDA",
    start: str = "2022-01-01",
    end: str = "2026-01-01",
) -> pd.DataFrame:
    """Return OHLCV dataframe.

    If Yahoo download fails or returns empty (common in locked-down networks),
    fall back to a deterministic synthetic random-walk series so the RL pipeline
    can still run end-to-end.
    """

    try:
        df = yf.download(ticker, start=start, end=end, progress=False)
    except Exception:
        df = pd.DataFrame()

    if df is None or df.empty:
        n = 800
        rng = np.random.default_rng(7)
        steps = rng.normal(loc=0.0005, scale=0.02, size=n)
        close = 100.0 * np.exp(np.cumsum(steps))
        idx = pd.date_range(start=start, periods=n, freq="B")
        df = pd.DataFrame(
            {
                "Open": close,
                "High": close * (1.0 + rng.uniform(0.0, 0.01, size=n)),
                "Low": close * (1.0 - rng.uniform(0.0, 0.01, size=n)),
                "Close": close,
                "Volume": rng.integers(1_000_000, 5_000_000, size=n),
            },
            index=idx,
        )

    return df


# 1. Get Historical Data
df = get_price_data()

# 2. Create the Environment
# window_size=10 means the agent sees last 10 days as its state
window_size = 10
frame_end = min(500, len(df) - 1)
if frame_end <= window_size:
    raise ValueError(f"Not enough data points: len(df)={len(df)}")

env = gym.make("stocks-v0", df=df, frame_bound=(window_size, frame_end), window_size=window_size)

# 3. Initialize the DQN Agent
model = DQN(
    "MlpPolicy",
    env,
    verbose=1,
    learning_rate=1e-3,
    buffer_size=50_000,
    learning_starts=1_000,
    target_update_interval=500,
    exploration_fraction=0.1,
    device="auto",
)

# 4. Train the Agent
print("Training started...")
model.learn(total_timesteps=20_000)
print("Training complete.")

# 5. Evaluate the Agent
print("\nTesting the trained agent...")
obs, info = env.reset()
done = False
while not done:
    action, _states = model.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = env.step(action)
    done = terminated or truncated

# 6. Visualize Results
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 6))
env.unwrapped.render_all()
plt.show()
```
